```python
import numpy as np
import scipy
```

# one-sample t-test

```python
x = [2.3, 2.5, 2.1, 2.7]

statistic, pvalue = scipy.stats.ttest_1samp(
    a=x, 
    popmean=2.0,
    alternative='two-sided', # ‘two-sided’, ‘less’, ‘greater’
)
print(statistic) # t statistic
print(pvalue)
```

# two-sample t-test

### independent (i.e. unpaired)

```python
x = [2.3, 2.5, 2.1]
y = [1.9, 2.0, 2.2]

statistic, pvalue = scipy.stats.ttest_ind(
    a=x, 
    b=y, 
    equal_var=False, # uses Welch test when equal_var=False
    alternative='two-sided', # ‘two-sided’, ‘less’, ‘greater’
)  
print(statistic) # t statistic
print(pvalue)
```

```python
# non-parametric alternative
statistic, pvalue = scipy.stats.mannwhitneyu(x, y, alternative='two-sided')
print(statistic)
print(pvalue)
```

### paired

```python
before = [2.1, 2.3, 2.5]
after  = [2.4, 2.5, 2.7]

statistic, pvalue = scipy.stats.ttest_rel(
    a=before, 
    b=after,
    alternative='two-sided', # ‘two-sided’, ‘less’, ‘greater’
)
print(statistic) # t statistic
print(pvalue)
```

```python
# non-parametric alternative
statistic, pvalue = scipy.stats.wilcoxon(before, after)
print(statistic)
print(pvalue)
```

# chi-squared test

```python
observed = np.array([[10,20],
                     [30,40]])

statistic, pvalue, dof, expected = scipy.stats.chi2_contingency(observed)
print(expected)
print(dof)
print(statistic) # X2
print(pvalue)
```

# fisher's exact test

```python
table = np.array([[1,9],
                  [11,3]])

statistic, pvalue = scipy.stats.fisher_exact(table, alternative='two-sided')
print(statistic) # odds ratio
print(pvalue)
```

# proportions z-test

```python
from statsmodels.stats.proportion import proportions_ztest

count = [30, 20]   # successes in treatment, control
nobs  = [50, 50]   # sample sizes

statistic, pvalue = proportions_ztest(count, nobs, alternative='larger')
print(statistic) # z-score
print(pvalue)
```

# correlation tests

```python
x = [2.3, 2.5, 2.1]
y = [1.9, 2.0, 2.2]

statistic, pvalue = scipy.stats.pearsonr(x, y) 
print(statistic) # r
print(pvalue)
```

```python
statistic, pvalue = scipy.stats.spearmanr(x, y)
print(statistic) # r
print(pvalue)
```
