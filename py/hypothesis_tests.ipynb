```python
import numpy as np
import scipy
```

# one-sample t-test

```python
x = [2.3, 2.5, 2.1, 2.7]

t_stat, p_value = scipy.stats.ttest_1samp(
    a=x, 
    popmean=2.0,
    alternative='two-sided', # ‘two-sided’, ‘less’, ‘greater’
)
print(t_stat)
print(p_value)
```

# two-sample t-test

### unpaired

```python
x = [2.3, 2.5, 2.1]
y = [1.9, 2.0, 2.2]

t_stat, p_value = scipy.stats.ttest_ind(
    a=x, 
    b=y, 
    equal_var=False, # uses Welch test when equal_var=False
    alternative='two-sided', # ‘two-sided’, ‘less’, ‘greater’
)  
print(t_stat)
print(p_value)
```

### paired

```python
before = [2.1, 2.3, 2.5]
after  = [2.4, 2.5, 2.7]

t_stat, p_value = scipy.stats.ttest_rel(
    a=before, 
    b=after,
    alternative='two-sided', # ‘two-sided’, ‘less’, ‘greater’
)
print(t_stat)
print(p_value)
```

# chi-squared test

```python
observed = np.array([[10,20],
                     [30,40]])

chi2_stat, p_value, dof, expected = scipy.stats.chi2_contingency(observed)
print(expected)
print(dof)
print(chi2_stat)
print(p_value)
```
