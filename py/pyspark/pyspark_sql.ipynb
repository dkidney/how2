```python
import re
import pyspark.sql.functions as f

def extract_words_from_document(document, regex):
    return re.findall(regex, document)

udf_func1 = f.udf(lambda document, regex: extract_words_from_document(document, regex), ArrayType(StringType()))
udf_func2 = f.udf(lambda document, regex: re.findall(regex, document), ArrayType(StringType()))

pattern = "(?:\\b[A-za-z'-]+\\b)"

distinct_urls\
    .withColumn('features1', udf_func1(f.col('text'), f.lit(pattern)))\
    .withColumn('features2', udf_func2(f.col('text'), f.lit(pattern)))\
    .show(5, True)
```
