```python
# pip install pyyaml

import yaml


def read_yaml_file_as_dict(path):
    # returns a dict
	with open(path, "r") as fh:
		try:
			return yaml.safe_load(fh)
		except yaml.YAMLError as exc:
			print(exc)


def convert_dict_to_yaml_string(a_dict):
	return yaml.dump(a_dict, fh, default_flow_style=False)


def write_dict_to_yaml_file(a_dict, path):
	with open(path, "w") as fh:
		yaml.dump(a_dict, fh, default_flow_style=False)


def print_dict_as_yaml_string(a_dict):
	print(yaml.dump(a_dict, default_flow_style=False))
```
